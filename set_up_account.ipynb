# Saving IBM Account to Qiskit

Get your API key from your IBM Quantum/IBM Cloud account, and paste it into `token` below. Your account will now be saved on your local computer, and will be used anytime you use Qiskit for something that requires an account.

```python
from qiskit_ibm_runtime import QiskitRuntimeService

token = 'your_API_key'
instance = 'your_instance_CRN' # OPTIONAL!

QiskitRuntimeService.save_account(token=token, channel="ibm_quantum_platform", instance=instance, overwrite=True)
```

```python
# Check that it worked
service = QiskitRuntimeService()
```
