```python
def decrypt(bitstring):
    bitstring = bitstring[0:36]
    chars = [' ','a','b','e','f','h','i','k','l','n','o','p','r','s','t','.']
    bitsplice = [bitstring[0+4*n:4+4*n] for n in range(0,9)]
    charstring = ""
    for sec in bitsplice:
        charstring += chars[(int(sec[3]) + 2*int(sec[2]) + 4*int(sec[1]) + 8*int(sec[0]))]
    return charstring

# decrypt takes the measured output bitstring (37 bits long) and outputs the corresponding secret message
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

oracle_circuit = QuantumCircuit(37)
oracle_circuit.x([1,2,3,5,7,9,11,13,14,16,17,18,19,21,24,25,26,31,34,35,36])
oracle_circuit.mcx(list(range(1,37)),0)
oracle_circuit.x([1,2,3,5,7,9,11,13,14,16,17,18,19,21,24,25,26,31,34,35,36])

oracle = oracle_circuit.to_gate(label="oracle")

# This is our oracle gate. It uses qubit 0 as the working qubit
```

```python
# Example of how to use the oracle gate

qc = QuantumCircuit(37)

qc.append(oracle,list(range(37)))

qc.measure_all()
```
